```
import numpy as np

import matplotlib.pyplot as plt


#spherical gaussian

#gausian distribution in two dimensions

r = np.random.rand(1000,2)

plt.scatter(r[:,0],r[:,1])


```

```
#lets look at cases where variance is different for each dimension
#lets put the SD for 2nd dimension to be 5 and mean to be 2

r[:,1] = 5*r[:,1]+2

plt.axis('equal')
plt.scatter(r[:,0],r[:,1])

plt.show()
```
